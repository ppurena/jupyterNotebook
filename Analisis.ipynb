```python
import pandas as pd
data = pd.read_csv('./Rfroom.csv', sep=',')
data.TIME_STAMP = pd.to_datetime(data.TIME_STAMP)
data.index = data.TIME_STAMP
data.drop(['TIME_STAMP'], axis=1, inplace=True)
data.describe(include= "all")
```

```python
data.corr()
```

```python
retr_stat = data.groupby("RETR_ANT").count()
retr_stat.LOST_PACKS
```

```python
pl = data.RETR_ANT.value_counts()[1:-1].plot(kind="bar",figsize=(20,10), legend = True)
pl.set_ylabel("Transmisiones")
pl.set_xlabel("Número de envios")
```

```python
data.RETR_ANT.value_counts()[3:-1].plot.pie(figsize=(6, 6))
```

```python
hours = pd.date_range(data.index[0], data.index[-1], freq="H", normalize=False)
dataHour = data.reindex(hours)
dataHour = dataHour.interpolate(method="nearest")
retrHr = dataHour.LOST_PACKS.diff()
retrHr = retrHr.dropna()
retrHr.count()
```

```python
retrHr.head(20)
```

```python
pl = retrHr.plot(figsize=(20,12), legend = True, grid = True)
pl.set_ylabel("Número de paquetes perdidos")
pl.set_xlabel("Horas del día, formato MM-DD-hh")
```

```python
unsampled_data = data.reindex(hours)
unsampled_data = unsampled_data.interpolate(method="nearest")
unsampled_data
```
